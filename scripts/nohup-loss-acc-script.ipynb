```python
import re
import pandas as pd
```

```python
csv_data = []
with open("nohup.out") as f:
    for line in f:
        data = []
#         x = re.match("[\w\W]* loss: (\w\W\w+) - acc: (\w\W\w+)", line) #loss , acc
        x = re.match("[\w\W]*step - loss: (\w\W\w+) - acc: (\w\W\w+)", line)
#         x = re.match("[\w\W]* - (\d+)s (\d+)[\w\W]*step - loss: (\w\W\w+) - acc: (\w\W\w+)", line)
        if x:
            for y in range(1, 3):
                data.append(x.group(y))
            csv_data.append(data)
# csv_data
```

```python
df = pd.DataFrame(csv_data, columns = ['loss', 'acc'])
```
