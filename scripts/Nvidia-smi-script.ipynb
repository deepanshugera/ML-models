```python
import re
import pandas as pd
```

```python
csv_data = []
with open("../nvidia.txt") as f:
    for line in f:
        data = []
        x = re.match("[\w\W]*N/A\s*(\d+)C\s*P0\s*(\d+)W[\w\W]*\s* (\d+)MiB /[\w\W]*\s* (\d+)%[\w\W]*", line)
        if x:
            for y in range(1,5):
                data.append(x.group(y))
            csv_data.append(data)
```

```python
df = pd.DataFrame(csv_data, columns = ['Temperature', 'Wattage', 'Memory', 'Utilization'])
```
