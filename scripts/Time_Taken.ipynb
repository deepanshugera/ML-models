```python
import re
import pandas as pd
```

```python
csv_data = []
with open("nohup.out") as f:
    for line in f:
        data = []
        x = re.match("[\w\W]* - (\d+)s (\d+)[\w\W]*step - loss: (\w\W\w+) - acc: (\w\W\w+)", line)
        if x:
            for y in range(1, 2):
                data.append(x.group(y))
            csv_data.append(data)
# csv_data
```

```python
df = pd.DataFrame(csv_data, columns = ['Time_Taken'])
```
